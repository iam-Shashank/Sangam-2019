```python
L1,L2=[],[]
for i in train.date_time:
    t1,t2=i.split(' ')
    L1.append(t1)
    L2.append(t2)
train['date']=L1
train['time']=L2
```

```python
train.info()
```

```python
day=1
L1=[]
L2=[]
L3=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
x='02'
for i in range(len(train)):
    if x!=train.iloc[i].date.split('-')[2]:
        day=(day+1)%7+1
        x=train.iloc[i].date.split('-')[2]
    else:
        day=day%7+1
    L1.append(day)
    L2.append(L3[day-1])
print(L1,L2)
```

```python
train['day']=L2
train['day_code']=L1
```
